```python
import pandas as pd
import re
#reading all the files in seprate days
file=pd.ExcelFile("timetable.xlsx",)
monday=file.parse(1,skiprows=[0,1,2,3],inplace=True)
tuesday=file.parse(2,skiprows=[0,1,2,3],inplace=True)
wednesday=file.parse(3,skiprows=[0,1,2,3],inplace=True)
thursday=file.parse(4,skiprows=[0,1,2,3],inplace=True)
friday=file.parse(5,skiprows=[0,1,2,3],inplace=True)
```

```python
#this will drop all nan coloumns
monday=monday.dropna(axis=1,how='all')
#tuesday
#this will drop all nan coloumns
tuesday=tuesday.dropna(axis=1,how='all')
#this will drop nan index
tuesday=tuesday.drop("Unnamed: 5",axis=1)
tuesday=tuesday.drop("Unnamed: 9",axis=1)
#wednesday
#this will drop all nan coloumns
wednesday=wednesday.dropna(axis=1,how='all')
wednesday=wednesday.drop("Unnamed: 14",axis=1)
#thursday
#this will drop all nan coloumns
thursday=thursday.dropna(axis=1,how='all')
thursday=thursday.drop("Unnamed: 5",axis=1)
thursday=thursday.drop("Unnamed: 9",axis=1)
#friday
friday=friday.dropna(axis=1,how='all')
friday=friday.drop("Unnamed: 14",axis=1)

```

```python
#setting up labs data Frame
#monday
monday_lab=monday.loc[36:]
monday_lab=monday_lab.reset_index(drop=True)
monday_lab.columns=monday_lab.iloc[0]
monday_lab=monday_lab.drop(0)
monday_lab=monday_lab.reset_index(drop=True)
monday_lab=monday_lab.dropna(axis=1,how='all')
monday_lab=monday_lab.fillna("Free")
monday.drop(monday.index[35:53],0,inplace=True)
# monday
#tuesday
tuesday_lab=tuesday.loc[36:]
tuesday_lab=tuesday_lab.reset_index(drop=True)
tuesday_lab.columns=tuesday_lab.iloc[0]
tuesday_lab=tuesday_lab.drop(0)
tuesday_lab=tuesday_lab.reset_index(drop=True)
tuesday_lab=tuesday_lab.dropna(axis=1,how='all')
tuesday.drop(tuesday.index[35:53],0,inplace=True)
tuesday_lab=tuesday_lab.fillna("Free")
# tuesday
# #wednesday
wednesday_lab=wednesday.loc[36:]
wednesday_lab.reset_index(drop=True,inplace=True)
wednesday_lab.columns=wednesday_lab.iloc[0]
wednesday_lab=wednesday_lab.drop(0)
wednesday_lab=wednesday_lab.reset_index(drop=True)
wednesday_lab.dropna(axis=1,how='all',inplace=True)
wednesday_lab.fillna("Free",inplace=True)
wednesday.drop(wednesday.index[35:53],0,inplace=True)
# wednesday
# #thursday
thursday_lab=thursday.loc[35:]
thursday_lab.reset_index(drop=True,inplace=True)
thursday_lab.columns=thursday_lab.iloc[0]
thursday_lab=thursday_lab.drop(0)
thursday_lab=thursday_lab.reset_index(drop=True)
thursday_lab.dropna(axis=1,how='all',inplace=True)
thursday_lab.fillna("Free",inplace=True)
thursday.drop(thursday.index[35:52],0,inplace=True)
#thursday
# #friday
friday_lab=friday.loc[33:]
friday_lab.reset_index(drop=True,inplace=True)
friday_lab.columns=friday_lab.iloc[0]
friday_lab=friday_lab.drop(0)
friday_lab=friday_lab.reset_index(drop=True)
friday_lab.dropna(axis=1,how='all',inplace=True)
friday.drop(friday.index[33:51],0,inplace=True)
friday_lab.fillna("Free",inplace=True)
# friday
```

# Finding all week free classes

```python
room=[]
ti=[]
day=[]
for k in range(0,len(classes)):
    classes[k]=classes[k].fillna("Free")
    temp=classes[k]
    temp=temp.fillna("Free")
    for i in range(len(temp.index)):
        for j in range(0,len(temp.columns)):
            subject=temp.iloc[i,j]
            col=temp.columns[j]
            if subject.find("Free") != -1 and col.find("Unnamed")== -1:
                time=temp.columns[j]
                day.append(days[k])
                room.append(temp.iloc[i,0])
                ti.append(time)
#                 print(days[k]," ",temp.iloc[i,0],"  ","  ",time)
freeDf=pd.DataFrame(day,columns=["Day"])
freeDf["Room"]=room
freeDf["Time"]=ti
freeDf
```

# Finding Less busy class in overall week

```python
room=[]
for i in range(0,len(classes[0].index)):
    room.append(classes[0].iloc[i,0])
df=pd.DataFrame(room,columns=["Room"])
df["Free Classes"]="0"
for k in range(0,len(classes)):
    classes[k]=classes[k].fillna("Free")
    temp=classes[k]
    temp=temp.fillna("Free")
    for i in range(len(temp.index)):
        for j in range(0,len(temp.columns)):
            subject=temp.iloc[i,j]
            col=temp.columns[j]
            if subject.find("Free") != -1 and col.find("Unnamed")== -1:
                number=df.iloc[i,1]
                number=int(number)
                number+=1
                df.iloc[i,1]=number
df=df.sort_values("Free Classes",ascending=False)
df
# free_class=df.iloc[0,0]
print("Most Less busy Class in overall week is:   ",df.iloc[0,0])
print("second less busy Class in overall week is: ",df.iloc[1,0])
```
